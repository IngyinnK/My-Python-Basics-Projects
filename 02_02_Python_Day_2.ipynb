```python
# Write a Python program to check if a number is positive, negative or zero.

num = float(input("Enter a number: "))

if num > 0:
    print("The number is Positive.")
elif num < 0:
    print("The number is Negative.")
else:
    print("The number is Zero.")
```

```python
# Write a Python program to check if a number entered by the user is odd or even.
# Odd numbers are not divisible by 2.
# Even numbers have remainder 0 when divided by 2.

num = int(input("Enter a number: "))

if num % 2 == 0:
    print("The number is Even.")
else:
    print("The number is Odd.")
```

```python
# Write a Python program to check if a number entered by the user is multiple of 3.

num = int(input("Enter a number: "))

if num % 3 == 0:
    print(f"{num} is a multiple of 3.")
else:
    print(f"{num} is not a multiple of 3.")
```

```python
# Write a Python program to check if a number entered by the user is multiple of 3 and 5.

num = int(input("Enter a number: "))

if num % 3 == 0 and num % 5 == 0:
    print(f"{num} is a multiple of both 3 and 5")
else:
    print(f"{num} is not a multiple of both 3 and 5")
```

# While Loop

```python
number = 1
while number <= 5:
  print(number)
  number +=1
```

```python
# car park

counter = 0
while counter <10:
  print("There is a space")
  counter +=1
  if counter == 10:
    print("Car park is full.")
```

```python
i = 1
while i<6:
  print(i)
  if i == 3:
    break
  i +=1
```

```python
password = "password"
user_entry = input("Enter the password: ")

if user_entry == password:
    print("Password is correct. Login successful.")

while user_entry != password:
    print("Password is incorrect. Login failed.")
    user_entry = input("Try again: ")

    if user_entry == password:
        print("Password is correct. Login successful.")
        break
```

```python
i = 1
while i <=10:
  if i % 2 == 0:
    print(f"{i} is an even number.")
  else:
    print(f"{i} is an odd number.")
  i +=1
```

```python
# Write a program to find and print multiples of 7 between 0 to 50.

i = 0
while i<=50:
  if i%7==0:
    print(f"{i} is multiple of 7.")
  i +=1
```

```python
secret_number = 8
tries_left = 3

while tries_left > 0:
    guess = int(input("Guess the secret number (1-10): "))

    if guess == secret_number:
        print("🎉 Correct! You win!")
        break

    elif guess < secret_number:
        print("Too low.")
    else:
        print("Too high.")

    tries_left -= 1
    print(f"Tries left: {tries_left}")

if tries_left == 0:
    print("😢 Game over. The correct number was", secret_number)
```

```python
# print all odd and even numbers between 1 and 10.

for i in range (1, 10): # stopping point (10) is not included.
  if i %2 == 0:
    print(f"{i} is even.")
  else:
    print(f"{i} is odd.")
```

```python
for i in range (1, 10, 2):
  print (i)

# range(start, stop, step)
# start number 1, stop before 10, increase by 2 each time
# So the loop prints numbers starting from 1, adding 2 each time, until it reaches 9.
```
