# Casting (type coversion)

```python
myage = "18"
print(type(myage))
# 18 is string.
```

```python
print(int(myage))
myage = int(myage)
# int function will convert age into numeric.
```

```python
print(myage*2)
```

```python
myage = float(myage)
print(myage)
# float function convert age into decimal.
```

```python
print("This is simple calculator.")

number1 = float(input("Enter first number: "))

number2 = float(input("Enter second number:"))

print(number1, "/", number2, "=", number1/number2)
```

```python
# float conversion here is safe because user entry can be integer or decimal.
# you can convert int to decimal but cannot convert decimal to int.
```
